```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

OnlineNewsPopularity = pd.read_csv("OnlineNewsPopularity.csv")
OnlineNewsPopularity
```

```python
from sklearn.model_selection import cross_val_score,train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    OnlineNewsPopularity.drop([' shares', 'url'], axis = 1),
    OnlineNewsPopularity[' shares'],
    test_size = 0.2,
    random_state = 42)
```

# Catboost with all predictors:

```python
#RMSE based on the parameter values that provide a cross validation R-squared close to the optimal one
model = CatBoostRegressor().fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
print("CatBoost RMSE = ",np.sqrt(mean_squared_error(model.predict(X_test),y_test)))
```

```python
from sklearn.metrics import r2_score
r2_score(model.predict(X_test), y_test)
```

# LightGBM with all predictors

```python
model_2 = LGBMRegressor().fit(X_train, y_train)
```

```python
print("LightGBM RMSE = ",np.sqrt(mean_squared_error(model_2.predict(X_test),y_test)))
```

```python
r2_score(model_2.predict(X_test), y_test)
```
